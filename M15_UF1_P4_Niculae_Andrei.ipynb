```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv("/home/cloudera/Desktop/london2018.csv")
print(weather)
```

First we create a new DataFrame in order to access our data file.

```python
weather.plot(y="Tmax", x="Month")
plt.show()
```

The grafic above shows the increase and decrease of the temperatures during the whole year.

```python
weather.plot(kind="line", y="Tmax", x="Month")
plt.show()
```

```python
weather.plot.line(y="Tmax", x="Month")
plt.show()
```

To specify that we want the chart to be a simple line chart, we can use either of the 2 examples above.

```python
weather.plot(y=['Tmax','Tmin'], x='Month')
plt.show()
```

This is an example of a line chart with more than one line. It is posible to add more lines to our chart, as long as the values in our data are aplicable to the chart. For example: adding Tmax and Rain into our chart wouldn't make sense because they represent 2 distinct values that don't coincide.

```python
weather['Tmed'] = (weather['Tmax'] + weather['Tmin'])/2
weather.plot(y=['Tmax','Tmin','Tmed'], x='Month')
plt.show()
```

Here we created a new column named Tmed, that represents the average temperature between Tmax and Tmin.

```python
weather.plot(kind='bar', y='Rain', x='Month')
plt.show()
```

Here we have an example of a bar chart. This chart shows the amount of rain from each month of the year. To have our chart be a bar chart or any other type, we simple use the "kind='...'" function, and specify witch type of chart we prefer.

```python
weather.plot(kind='barh', y='Rain', x='Month')
plt.show()
```

'barh' is use for showing a bar chart horizontaly.

```python
weather.plot(kind='bar', y=['Tmax','Tmin'], x='Month')
plt.show()
```

```python
weather.plot(kind='bar', y=['Tmax','Tmed','Tmin'], x='Month')
plt.show()
```

```python
weather.plot(kind='scatter', x='Sun', y='Rain')
plt.show()
```

A scatter diagram (as the one seen above) plots a series of points that correspond to two variables and allows us
to determine if there is a relationship between them.

```python
weather.plot(kind='pie', y='Sun')
plt.show()
```

Pie charts are typically used to show what proportion of some value can be associated
with a particular group or category.

```python
weather.index=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
weather.plot(kind='pie', y = 'Sun', legend=False)
plt.show()
```

Here we got replaced the numbers (witch represented the index, the numbers in the first column ) with the name of their respective month to have an easier time understanding the chart. We also got rid of the legend because it was hiding a small part of the chart.

Here we have a basic overview of what our DataSet.

```python
more_weather.plot.box(y='Rain')
plt.show()
```

In this chart the data is divided into 3 parts: 
- the center box represents the halfway point through the data, there are 50% data points above and below it, it is also classified as Q2
- the bottom line reprsents the first 25%, everything below 25% is classified as Q1
- the top line represents the last 25%, everything above 75% is classified as Q3
The centre line is the second quartile(Q2), and is also the median of the data.
The bottom of box is the first quartile(Q1) and is also the median of the bottom half of the data.
The top of the box is the third quartile(Q3), and is also the the median of the top half of the data.
The whiskers extend to 1.5 times inter quartile range(IQR), which is Q3-Q1, from edge of box.
The circles are outliers, the individual values that lie beyond the end of the whiskers.
