```python
#exrciseC
#returns the top 15 movies regarding the score on imdb
topscore=movies.sort_values('imdb_score',ascending=False)
topscore2=topscore[['movie_title','title_year','genres','imdb_score']]
topscore2.head(15)
```

```python
#exerciseD
#returns the top 10 movies regsarding the net profit
movies['net_profit']=movies['gross'] - movies['budget']
movies['net_profit'].head()
mostprofit=movies.sort_values('net_profit',ascending=False)
mostprofit[['movie_title','genres','net_profit']].head(10)
```

```python
#returns the top 10 movies regarding the percentage profit
movies['profit_percen']=((movies['gross'] - movies['budget']) / movies['budget']) * 100
movies['profit_percen'].head()
mostpercentage=movies.sort_values('profit_percen',ascending=False)
mostpercentage[['movie_title','genres','profit_percen']].head(10)
```

```python
#finalExercise
#returns how many movies each director has made during the last 26 years
totalM=movies[movies['title_year']>1994]
totalM['director_name'].value_counts().head(20)
```
